```python
import numpy as np
import pandas as pd
import csv
```

```python
df = pd.read_csv("employment/employment_data_by_county.csv")
df.head()
```

```python
legend_df = pd.read_csv("employment/employment_legend.csv")
legend_df
```

```python
df_estimate = df.loc[:, ~df.columns.str.endswith('M')]
df_estimate.head()
```

```python
df_sector = df_estimate.loc[:, df_estimate.columns.str.contains("C01")]
df_sector.head()
```

```python
df_sector.shape
```

```python
sector_name = legend_df[legend_df["GEO_ID"].str.contains("C01")]
sector_name
```

```python
sector_name = sector_name[sector_name["GEO_ID"].str.endswith("E")]
sector_name
```

```python
sector_name.shape
```

```python
sector_name_series = sector_name["id"].to_list()
sector_name_series
```

```python
df_sector = df_sector.drop(["S2405_C01_001E", "S2405_C01_015E"],1)
df_sector.head()
```

```python
df_sector = df_sector.drop(df_sector.index[0])
df_sector.head()
```

```python
sector_name['id'] = sector_name['id'].map(lambda x: x.lstrip('Estimate!!Total!!Civilian employed population 16 years and over!!'))
sector_name
```

```python
sector_name = sector_name.drop([sector_name.index[0], sector_name.index[14]])
sector_name = sector_name.replace("ction", "Construction")
sector_name
```

```python
ID_sequence = sector_name["GEO_ID"]
ID_sequence
```

```python
df_sector = df_sector[ID_sequence]
df_sector.head()
```

```python
sectors = sector_name["id"]
```

```python
df_sector.columns = sectors
df_sector.head()
```

```python
max_sector = df_sector.max(axis = 1)
max_sector
```

```python
df_sector.dtypes
```

```python
df_sector = df_sector.apply(pd.to_numeric) 
df_sector.dtypes
```

```python
max_sector = df_sector.max(axis = 1)
max_sector
```

```python
df_sector["Largest_Sector"] = max_sector
df_sector.head()
```

```python
df_largest_sector = df_sector.filter(["Largest_Sector"], axis = 1)
df_largest_sector.head()
```

```python
df_largest_sector.shape
```

```python
df_largest_sector = df_sector.idxmax(axis=1)
df_largest_sector.head()
```

```python
df_sector["Largest_sector"] = df_largest_sector
df_sector.head()
```

```python
df_sector = df_sector.replace("cational services, and health care and social assistance",
                              "Educational services, and health care and social assistance")
df_sector.head()
```

```python
df_sector = df_sector.replace("warehousing, and utilities",
                              "Transportation and warehousing, and utilities")
df_sector.head()
```

```python
df_clean_sector = df_sector.filter(["Largest_sector"], axis = 1)
df_clean_sector.head()
```

```python
df_clean_sector.to_csv('largest_sector_by_county.csv')
```
